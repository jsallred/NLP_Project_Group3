```python
import pandas  as pd
import numpy as np
import os
from collections import Counter
```

```python
filepath:str = './scraped_data/politicalBias_scraped.csv'
df = pd.read_csv(filepath)
print(type(df))
len(df)
```

```python
publishers = []
for url in df['url']:
    publisher = url.split('.')[1]
    publishers.append(publisher)
```

```python
from collections import Counter
unique_pubs = {}

unique_pubs = Counter(publishers)
print(unique_pubs)
```

```python
fox_data = pd.DataFrame()
nytimes_data = pd.DataFrame()
washingtontimes_data = pd.DataFrame()
washingtonpost_data = pd.DataFrame()
npr_data = pd.DataFrame()

for index,row in df.iterrows():
    if( 'www.foxnews.com' in row['url']):
        fox_data = fox_data._append(row)
    if( 'www.nytimes.com' in row['url']):
        nytimes_data = nytimes_data._append(row)
    if( 'www.washingtontimes.com' in row['url']):
        washingtontimes_data = washingtontimes_data._append(row)
    if( 'www.washingtonpost.com' in row['url']):
        washingtonpost_data = washingtonpost_data._append(row)
    if( 'www.npr.org' in row['url']):
        npr_data = npr_data._append(row)
        
fox_data.drop(columns='Unnamed: 0',axis=1,inplace=True)

fox_data.reset_index(drop = True,inplace=True)
print(fox_data.head())
    
```

```python
def relable_class(row): # classify 
    if row['label'] == 'Left':
        return 0
    elif row['label'] == 'Center':
        return 1
    elif row['label'] == 'Right':
        return 2
    else:
        return row['label']  # Return the original value if not one of the specified types
```

```python
fox_data['label'] = fox_data.apply(relable_class, axis=1)
npr_data['label'] = npr_data.apply(relable_class, axis=1)
nytimes_data['label'] = nytimes_data.apply(relable_class, axis=1)
washingtonpost_data['label'] = washingtonpost_data.apply(relable_class, axis=1)
washingtontimes_data['label'] = washingtontimes_data.apply(relable_class, axis=1)
```

```python

fox_data.to_csv(f'./top_publishers/fox_data.csv')
npr_data.to_csv(f'./top_publishers/npr_data.csv')
nytimes_data.to_csv(f'./top_publishers/nytimes_data.csv')
washingtonpost_data.to_csv(f'./top_publishers/washingtonpost_data.csv')
washingtontimes_data.to_csv(f'./top_publishers/washingtontimes_data.csv')
```
